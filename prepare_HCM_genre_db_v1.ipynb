```python
import mirdata
from tqdm import tqdm
import os
import requests
import os
from pathlib import Path
from pydub import AudioSegment
import shutil
import numpy as np
import librosa
import soundfile as sf
import random
import json
genre_dir = Path(os.path.expanduser("/home/ganesh/fubar/music_agent_v1/genre_db_v1/"))
genre_dir.mkdir(exist_ok=True)

HCM_dataset = mirdata.initialize('saraga_hindustani',data_home='/home/ganesh/fubar/music_agent_v1/genre_db_v1')

if os.path.exists('/home/ganesh/fubar/music_agent_v1/genre_db_v1/saraga1.5_hindustani'):
    print("HCM dataset already exists.")

    mp3_files = list(genre_dir.rglob("*.mp3"))
    print(f"Found {len(mp3_files)} MP3 files")
else:
    HCM_dataset.download()
   
HCM_dataset.validate()
track = HCM_dataset.choice_track()  # load a random track
print(track)  # see what data a track contains
```

```python
# Build raag database from your Saraga dataset
def build_raag_database(dataset):
    raag_db = {}
    
    for track_id in dataset.track_ids:
        track = dataset.track(track_id)
        
        # Get raag name from metadata
        try:
            raag_name = track.metadata['raags'][0]['common_name']
            raag_name = raag_name.lower().strip()  # Normalize
        except:
            continue
        
        if raag_name not in raag_db:
            raag_db[raag_name] = []
        
        raag_db[raag_name].append(track.audio_path)
    
    return raag_db

# Build once
raag_db = build_raag_database(HCM_dataset)

# Save it
with open(genre_dir/'raag_lookup.json', 'w') as f:
    print   ("Saving raag database...")
    json.dump(raag_db, f, indent=2)

print(f"Available raagas: {len(raag_db)}")
for raag, clips in raag_db.items():
    print(f"  {raag}: {len(clips)} clips")
```
